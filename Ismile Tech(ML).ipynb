```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent=apriori(R,min_support=0.007,use_colnames=True)
rules=association_rules(frequent,metric="lift",min_threshold=1)
rules
```

```python
records = []
for i in range(0, 2182):
    records.append([str(df.vales[i,j] for j in range(0, 80))])
```

```python
rules = apriori(records, min_support = 0.0045, min_confidence = 0.002, min_lift = 2, min_length = 2)
results = list(rules)
```
